```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('survey_results_public.csv')

df.head()
```

```python
df.info()
```

```python

df.describe(include = 'object')
```

```python
df['Employment'].nunique()
```

```python
df_employ = df['Employment']

employment = df_employ.value_counts()
top = employment.head(5)

plt.figure(figsize = (8,6))
top.plot.bar()

plt.xlabel('Employment')
plt.ylabel('Counts')

plt.xticks(rotation=45, ha='right')
# plt.tight_layout()

plt.show()
```
