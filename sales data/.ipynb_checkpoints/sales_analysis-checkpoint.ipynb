### Merge 12 months of sales data into one

```python
# data = pd.read_csv('Sales_Data/Sales_April_2019.csv')

# data.head()

# from gpt, it works

# csv_files = glob.glob('Sales_Data/Sales_*.csv')
# print(csv_files)

# dataframes = []

# for file in csv_files:
#     df = pd.read_csv(file)
#     dataframes.append(df)
    
# merged_df = pd.concat(dataframes, ignore_index=True)
# merged_df.to_csv('merged_data.csv', index=False)

# merged_df.tail()


# from tutorial

path = "./Sales_Data"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files

df = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    df = pd.concat([df, current_data])
    
df.to_csv("merged_data2.csv", index=False)
```

### Merged Dataframe

```python
df = pd.read_csv('merged_data2.csv')
# df = pd.read_csv('merged_data.csv')
df.head()
```

### Data Cleaning

```python
# df.loc[df['Order Date'].isnull()]

# nan_df = df[df.isnull().any(axis=1)]

# nan_df.head()

# drop nan values
df = df.dropna(how='all')

# Remove rows with placeholder values
df = df[df['Order Date'] != 'Order Date'] 

print(df.isnull().sum())
df.head()
```

### Add month and sales column, visualize by monthly sales

```python

# df['Month'] = df['Order Date'].dt.month

# df['Month'] = 4

# df['Month'] = df['Order Date'].str[0:2]
# df['Month'] = df['Order Date'].astype('int64')

# reset index after setting
# df.reset_index(inplace=True)

# find specific element using loc and iloc
# df.loc[(df['Order ID']== '177057')]
# print(df.iloc[519])


df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%y %H:%M')

df['Month'] = pd.to_datetime(df['Order Date']).dt.month
df['Month'] = df['Order Date'].dt.strftime('%B')
# df.drop('Month', axis =1)

#also works
# df['Quantity Ordered'] = df['Quantity Ordered'].astype(float)
# df['Price Each'] = df['Price Each'].astype(float)

df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])

df['Sales'] = df['Quantity Ordered'] * df['Price Each']

# df.info()
# df.head()

# arrange month order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September',
              'October','November','December']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)

monthly_sales = df.groupby('Month')['Sales'].sum()

fig, ax = plt.subplots(figsize=(8, 6))

# Plot the monthly sales data
ax.bar(monthly_sales.index, monthly_sales.values)
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.set_title('Monthly Sales')

# Customize the appearance of the plot
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--')

# Adjust layout and spacing
plt.tight_layout()

# Display the plot
plt.show()

# for double checking data (IMPORTANT)
print(monthly_sales)
```

### Add city column, sales by city, visualize

```python

# str.strip() is not necessary, but is for removing leading or trailing whitespace
# df['City'] = df['Purchase Address'].str.split(',').str[1].str.strip()

# drop multiple columns
# df.drop(columns = 'CityState', inplace=True)
# df.drop(['City','State'], axis=1, inplace=True)

df['City'] = df['Purchase Address'].str.split(',').str[1:3].apply(lambda x: ', '.join([s.strip() for s in x]))

# remove postal code
df['City'] = df['City'].str.replace(r'\b\d+\b', '', regex=True).str.strip()

# since my index messed [0,2,3,4,5] 
# reset index after dropping rows
# df = df.reset_index(drop=True)


# visualizing
sales_by_city = df.groupby('City')['Sales'].sum()

fig, ax = plt.subplots(figsize=(8, 6))

# Plot the data
ax.bar(sales_by_city.index, sales_by_city.values)
ax.set_xlabel('City')
ax.set_ylabel('Sales')
ax.set_title('Sales by City')

# Customize the appearance of the plot, rotate city names 45 degrees
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--')

# Adjust layout and spacing
plt.tight_layout()

# Display the plot
plt.show()

# for double checking data (IMPORTANT)
print(sales_by_city)
```
